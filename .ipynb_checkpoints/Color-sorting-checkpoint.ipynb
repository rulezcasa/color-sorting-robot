## Realtime tagging 

```python
import cv2


model_trained = YOLO('/Users/casarulez/Projects/color-sorting/best.pt')


camera = cv2.VideoCapture(0)  # Use 0 for default camera, you can change it according to your camera index

while True:

    ret, frame = camera.read()
    
    if not ret:
        print("Failed to capture frame")
        break


    output = model_trained([frame])


    for result in output:
        boxes = result.boxes  
        

        if boxes:

            box = boxes.xyxy[0].tolist() 
            x_min, y_min, x_max, y_max = box
            
         
            cv2.rectangle(frame, (int(x_min), int(y_min)), (int(x_max), int(y_max)), (0, 255, 0), 2)  # Draw bounding box

    # Display the frame with bounding box using OpenCV
    cv2.imshow('Object Detection', frame)

    # Check for key press to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera
camera.release()
cv2.destroyAllWindows()
```

## Integrating HSV parameters 

## Tackling Different Lighting Conditions with HSV Parameters

Integrating HSV parameters is a powerful technique for robust color detection, especially in scenarios with varying lighting conditions. HSV represents colors in a way that separates hue, saturation, and value, making it more robust to changes in brightness and illumination.

### Parameters Used

#### Red Color:
- **Lower Bound:** [0, 100, 100]
- **Upper Bound:** [10, 255, 255]

#### Blue Color:
- **Lower Bound:** [100, 100, 100]
- **Upper Bound:** [130, 255, 255]

#### Green Color:
- **Lower Bound:** [40, 100, 100]
- **Upper Bound:** [80, 255, 255]

### Explanation

- **Hue (H):** Represents the color type. By setting specific ranges for each color, variations due to lighting changes are minimized.
- **Saturation (S):** Reflects the intensity or purity of the color. By keeping saturation within a certain range, the detection remains consistent despite changes in lighting intensity.
- **Value (V):** Corresponds to the brightness of the color. By adjusting the value range, the detection adapts to different lighting conditions without being overly sensitive to brightness changes.

By tuning these HSV parameters, the color detection algorithm becomes more resilient to different lighting conditions, ensuring accurate and reliable performance in various environments.


```python
import cv2
import numpy as np


model_trained = YOLO('/Users/casarulez/Projects/color-sorting/best.pt')


camera = cv2.VideoCapture(0) 

while True:

    ret, frame = camera.read()
    
    if not ret:
        print("Failed to capture frame")
        break


    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    

    lower_red = np.array([0, 100, 100])
    upper_red = np.array([10, 255, 255])

    lower_blue = np.array([100, 100, 100])
    upper_blue = np.array([130, 255, 255])


    lower_green = np.array([40, 100, 100])
    upper_green = np.array([80, 255, 255])


    mask_red = cv2.inRange(frame_hsv, lower_red, upper_red)
    mask_blue = cv2.inRange(frame_hsv, lower_blue, upper_blue)
    mask_green = cv2.inRange(frame_hsv, lower_green, upper_green)
    

    res_red = cv2.bitwise_and(frame, frame, mask=mask_red)
    res_blue = cv2.bitwise_and(frame, frame, mask=mask_blue)
    res_green = cv2.bitwise_and(frame, frame, mask=mask_green)
    

    output = model_trained([frame])


    for result in output:
        boxes = result.boxes  
        

        if boxes:

            box = boxes.xyxy[0].tolist()  
            x_min, y_min, x_max, y_max = box
            

            cv2.rectangle(frame, (int(x_min), int(y_min)), (int(x_max), int(y_max)), (0, 255, 0), 2)  # Draw bounding box


    cv2.imshow('Object Detection', frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


camera.release()
cv2.destroyAllWindows()
```
